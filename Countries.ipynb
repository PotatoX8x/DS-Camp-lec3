```python
import pandas as pd
import numpy as np
from functools import reduce
```

```python
#creating main dataframe of 15 ranked countries
def answer_one():
    energy = pd.read_excel('Energy Indicators.xls', skiprows=17, skipfooter=38)
    energy = energy.drop(energy.columns[[0, 1]], axis=1)
    energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
    energy = energy.replace('...', np.nan)
    energy['Energy Supply'] *= 1000000
    energy['Country'] = energy['Country'].str.replace('\d+', '', regex = True)
    energy['Country'] = energy['Country'].str.replace(r'\(.*\)','', regex = True)
    energy['Country'] = energy['Country'].replace('Republic of Korea', 'South Korea')
    energy['Country'] = energy['Country'].replace('United States of America', 'United States')
    energy['Country'] = energy['Country'].replace('United Kingdom of Great Britain and Northern Ireland', 'United Kingdom')
    energy['Country'] = energy['Country'].replace('China, Hong Kong Special Administrative Region', 'Hong Kong')
    
    GDP = pd.read_csv('world_bank.csv', skiprows=4)
    GDP = GDP.rename(columns={'Country Name': 'Country'})
    GDP = GDP.drop(GDP.iloc[:, 1 : 50], axis=1)
    GDP = GDP.drop(GDP.iloc[:, 11:], axis=1)
    GDP['Country'] = GDP['Country'].replace('Korea, Rep.', 'South Korea')
    GDP['Country'] = GDP['Country'].replace('Iran, Islamic Rep.', 'Iran')
    GDP['Country'] = GDP['Country'].replace('Hong Kong SAR, China', 'Hong Kong')
    
    ScimEn = pd.read_excel('scimagojr country rank.xlsx')
    ScimEn = ScimEn.drop(['Region'], axis = 1)
    #merging
    data_frames = [ScimEn, energy, GDP]
    Top15 = reduce(lambda  left, right: pd.merge(left, right, on=['Country'], how='outer'), data_frames)
    Top15.set_index('Country', inplace = True)
    Top15 = Top15[Top15['Rank'] <= 15]
    return Top15
display(answer_one())
```

```python
#average GDP sorted in descendang order
def answer_two():
    Top15 = answer_one()
    avgGDP = pd.Series(index = Top15.index, dtype = 'float64')
    avgGDP = Top15.iloc[:, 10:].mean(axis = 1)
    avgGDP.sort_values(ascending = False, inplace = True)
    return avgGDP
display(answer_two())
```

```python
#6th largest average GDP country, how much GDP changed from 2006 to 2015
def answer_three():
    Top15_awGDP = answer_two()
    Top6_name = Top15_awGDP.index[5]
    Top15 = answer_one()
    change = Top15.loc[Top6_name]['2015'] - Top15.loc[Top6_name]['2006']
    return change
print(answer_three())
```

```python
#highest Self-citation/citations ratio
def answer_four():
    Top15 = answer_one()
    Top15.insert(6,'Self-citation/citations', 0)
    Top15['Self-citation/citations'] = Top15['Self-citations'] / Top15['Citations']
    Top15['Self-citation/citations'] = pd.to_numeric(Top15['Self-citation/citations'])
    max_row = Top15.loc[Top15['Self-citation/citations'].idxmax()]
    max_name = [max_row.name, max_row.at['Self-citation/citations']]
    return max_name
print(answer_four())
```

```python
#country with 3rd highest estimated population
def answer_five():
    Top15 = answer_one()
    Top15.insert(7,'Population', 0)
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['Population'] = Top15['Population'].fillna(0)
    Top3 = Top15['Population'].nlargest(3)
    return Top3.index[2]
display(answer_five())
```

```python
#correlation between Citable documents per capita and Energy Supply per Capita
def answer_six():
    Top15 = answer_one()
    Top15.insert(7,'Population', 0)
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15.insert(8,'Citable documents per capita', 0)
    Top15['Citable documents per capita'] = Top15['Citable documents'] / Top15['Population']
    corr_capita = Top15['Citable documents per capita'].corr(Top15['Energy Supply per Capita'])
    return corr_capita
display(answer_six())
```

```python
#Statistics on continents and population
def answer_seven():
    Top15 = answer_one()
    Top15.insert(7,'Population', 0)
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    stats = pd.DataFrame(index = ['Asia', 'Australia', 'Europe', 'North America', 'South America'], 
                         columns = ['size', 'sum', 'mean', 'std'])
    stats.index.name = 'Continent'
    #sorting given dict
    sorted_dict = {}
    for i, v in ContinentDict.items():
        sorted_dict[v] = [i] if v not in sorted_dict.keys() else sorted_dict[v] + [i]
    Cont_df = pd.DataFrame.from_dict({ key:pd.Series(value) for key, value in sorted_dict.items() })
    size_ser = Cont_df.count()
    stats['size'] = size_ser
    #filling series of population values for each continent
    data_Asia = Top15.loc[Cont_df['Asia'].dropna(), 'Population']
    data_NA = Top15.loc[Cont_df['North America'].dropna(), 'Population']
    data_SA = Top15.loc[Cont_df['South America'].dropna(), 'Population']
    data_Au = Top15.loc[Cont_df['Australia'].dropna(), 'Population']
    data_Eu = Top15.loc[Cont_df['Europe'].dropna(), 'Population']
    #sum
    stats.at['Asia', 'sum'] = data_Asia.sum()
    stats.at['North America', 'sum'] = data_NA.sum()
    stats.at['South America', 'sum'] = data_SA.sum()
    stats.at['Australia', 'sum'] = data_Au.sum()
    stats.at['Europe', 'sum'] = data_Eu.sum()
    #mean
    stats.at['Asia', 'mean'] = data_Asia.mean()
    stats.at['North America', 'mean'] = data_NA.mean()
    stats.at['South America', 'mean'] = data_SA.mean()
    stats.at['Australia', 'mean'] = data_Au.mean()
    stats.at['Europe', 'mean'] = data_Eu.mean()
    #std
    stats.at['Asia', 'std'] = data_Asia.std()
    stats.at['North America', 'std'] = data_NA.std()
    stats.at['South America', 'std'] = data_SA.std()
    stats.at['Australia', 'std'] = data_Au.std()
    stats.at['Europe', 'std'] = data_Eu.std()
    return stats
display(answer_seven())
```
